# Library


```python
class Book:
    def __init__(self, title, author, isbn):
        # Initialize the attributes
        self.title = title
        self.author = author
        self.isbn = isbn
        self.available_copies = 0
        self.total_copies = 0

    def borrow_book(self, num_copies):
        #  Fill in the blank: Update the number of available copies when a book is borrowed
        if num_copies <= self.available_copies:
            self.available_copies -= num_copies

        else:
            print("sorry, insufficient quantity")
            # Ensure that the number of copies borrowed does not exceed the available copies

    def return_book(self, num_copies):
        #  Fill in the blank: Update the number of available copies when a book is returned
        if (
            num_copies <= self.available_copies
            and num_copies > 0
            and num_copies <= (self.total_copies - self.available_copies)
        ):
            self.available_copies += num_copies
        else:
            print("sorry, insufficient quantity")
        # Ensure that the number of copies returned does not exceed the total copies

    def display_info(self):
        #  Fill in the blank: Print information about the book, including title, author, ISBN,
        # and the number of available copies
        print(
            f"data book:\n Title: {self.title},\n Author: {self.author},\n ISBN: {self.isbn},\n Available Copies: {self.available_copies},\n  Total Copies: {self.total_copies}"
        )


class Library:
    def __init__(self):
        # Initialize the library with an empty list of books
        self.books = []

    def add_book(self, book):
        #  Fill in the blank: Add a book to the library
        self.books.append(book)
        pass

    def search_book(self, title):
        search_result = []
        #  Fill in the blank: Search for a book in the library by title
        for book in self.books:
            if book.title.lower() == title.lower():
                search_result.append(book)
        return search_result

    def display_books(self):
        # Fill in the blank: Display information about all books in the library
        for id, book in enumerate(self.books):
            print(
                f"data library #{id}:\n title:{book.title},\n  author:{book.author},\n  isbn:{book.isbn},\n Available Copies: {book.available_copies},\n  Total Copies: {book.total_copies}"
            )

    @staticmethod
    def display_match_books(books):
        if len(books) > 0:
            for book in books:
                print(f"found the book {book.__dict__}")
        else:
            print("not found book with the title given.")


def main():
    # call class:
    library = Library()

    book1 = Book(
        title="The Catcher in the Rye", author="J.D. Salinger", isbn="978-0-316-76948-0"
    )
    book1.total_copies = 5
    book1.available_copies = 5

    book2 = Book(
        title="To Kill a Mockingbird", author="Harper Lee", isbn="978-0-06-112008-4"
    )
    book2.total_copies = 3
    book2.available_copies = 3

    # Add books to the library
    # Fill in the blanks:
    library.add_book(book1)
    library.add_book(book2)

    # Display all books in the library
    # Fill in the blank:
    library.display_books()

    # Search book
    search_1 = library.search_book("The Catcher in the Rye")
    library.display_match_books(search_1)
    search_2 = library.search_book("I Want to Die But I Want to Eat Tteokbokki")
    library.display_match_books(search_2)

    # Borrow a few copies of a book
    # Fill in the blank:
    book1.borrow_book(2)

    # Display updated information about the borrowed book
    # Fill in the blank:
    book1.display_info()

    # Return some copies of the borrowed book
    # Fill in the blank:
    book1.return_book(1)

    # Display updated information about the returned book
    # Fill in the blank:
    book1.display_info()


main()
```

```python
# Create a banking system with multiple account types: SavingsAccount, CheckingAccount, and BusinessAccount.
# Implement the following features:
# - Each account should have a unique account number.
# - The accounts should support deposit and withdrawal operations.
# - The SavingsAccount should have an interest rate, and interest should be calculated and added monthly.
# - The CheckingAccount should have an overdraft limit.
# - The BusinessAccount should have a credit limit.
# - Create a Bank class that manages a list of accounts and provides a method to display account information.

# Implement the classes and methods with appropriate encapsulation and inheritance.

# You can create instances of these classes and demonstrate their functionality.

# Note: This is a more advanced exercise, so take your time and think through the design.

from abc import abstractmethod
import random
from datetime import datetime, timedelta


class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        # Implement deposit method
        if amount > 0:
            self.balance += amount
            print(
                f"deposit into account with amount {self.balance} after deposit with amount {amount}"
            )
        else:
            print("deposit amount should more than 0")

    def add_interest_rate(self, interest_rate):
        # Implement add_interest_rate method
        self.balance = self.balance + (self.balance * interest_rate)

    @abstractmethod
    def withdraw(self, amount):
        pass


class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance=0, interest_rate=0.02):
        # Implement __init__ method
        self.account_number = account_number
        self.balance = balance
        self.interest_rate = interest_rate

    def calculate_interest(self):
        # Implement calculate_interest method
        self.balance = self.balance + (self.balance * self.interest_rate)
        return self.balance


class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance=0, overdraft_limit=100):
        # Implement __init__ method
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        # Implement withdraw method
        if (
            amount > 0
            and amount < self.balance
            and (self.balance - amount >= self.overdraft_limit)
        ):
            self.balance -= amount
            print(
                f"withdraw into account with amount {amount}, now yout balance {self.balance}"
            )
        else:
            print(f"invalid withdraw amount, exceed overdraft limit")


class BusinessAccount(BankAccount):
    def __init__(self, account_number, balance=0, credit_limit=500):
        # Implement __init__ method
        super().__init__(account_number, balance)
        self.credit_limit = credit_limit

    def withdraw(self, amount):
        # Implement withdraw method for BusinessAccount
        if (
            amount > 0
            and amount <= self.balance
            and (self.balance - amount >= -self.credit_limit)
        ):
            self.balance -= amount
            print(
                f"withdraw into account with amount {amount}, now yout balance {self.balance}"
            )
        else:
            print("invalid withdraw amount, exceed credit limit")


class Bank:
    def __init__(self):
        self.accounts = []

    def add_account(self, account):
        # Implement add_account method
        if account not in self.accounts:
            self.accounts.append(account)
            print(
                f"account number {account.account_number} added, account list now {account.__dict__}"
            )
        else:
            print("account already exist")

    def display_accounts_info(self):
        # Implement display_accounts_info method
        for account in self.accounts:
            print(f"Account {account.__dict__}")


# Example Usage:
bank = Bank()

# Create accounts
savings_account = SavingsAccount(
    account_number=random.randint(1000, 9999), balance=1000
)
checking_account = CheckingAccount(
    account_number=random.randint(1000, 9999), balance=500, overdraft_limit=200
)
business_account = BusinessAccount(
    account_number=random.randint(1000, 9999), balance=2000, credit_limit=1000
)
bank_account = BankAccount(account_number=random.randint(1000, 9999), balance=1000)
print("~~~~~~~~~~~~add_account~~~~~~~~~~~~")
# Add accounts to the bank
bank.add_account(savings_account)
bank.add_account(checking_account)
bank.add_account(business_account)

# Perform transactions
# # Perform transactions (e.g., deposit, withdraw, calculate interest)
print("~~~~~~~~~~~~savings_account~~~~~~~~~~~~")
savings_account.deposit(200)
savings_account.add_interest_rate(0.02)
print("~~~~~~~~~~~~checking_account~~~~~~~~~~~~")
checking_account.deposit(1000)
checking_account.withdraw(200)
print("~~~~~~~~~~~~business_account~~~~~~~~~~~~")
business_account.deposit(500)
business_account.withdraw(1000)
print("~~~~~~~~~~~~display account~~~~~~~~~~~~")
# Display account information
bank.display_accounts_info()
print("~~~~~~~~~~~~account duplication~~~~~~~~~~~~")
bank.add_account(savings_account)
bank.add_account(checking_account)
bank.add_account(business_account)
print("~~~~~~~~~~~~overdraft limit~~~~~~~~~~~~")
checking_account.withdraw(1000000)
print("~~~~~~~~~~~~credit limit~~~~~~~~~~~~")
business_account.withdraw(1000000)
```

# Task Encapsulation

```python
from abc import ABC, abstractmethod
import threading
import math
import time


# base class
class Shape(ABC):
    # class attribute only can be call on this class, not by the child class
    default_color = "black"

    def __init__(self, color=None):
        self.color = color or self.default_color

    # override by child class (Circle)
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    # staticmethod for calling without make the object first
    @staticmethod
    def is_valid_color(color):
        valid_colors = ["red", "green", "blue"]
        if color.lower() in valid_colors:
            print("valid color")
        else:
            print("not a valid color")

    def display_info_shape(self):
        print(f"Area: {self.area():.2f}")
        print(f"Perimeter: {self.perimeter():.2f}")


# Circle class inheriting from Shape
class Circle(Shape):
    def __init__(self, radius, color="black"):
        super().__init__(color)
        # private attribute
        self._radius = radius

    # only class circle can get the radius
    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, radius):
        self._radius = radius

    def area(self):
        return math.pi * self._radius**2

    def perimeter(self):
        return 2 * math.pi * self._radius

    # classmethod to call class attribute and method, except self
    @classmethod
    def display_info_color(cls):
        cls.is_valid_color("pink")
        print(f"default color: {cls.default_color}")

    def run(self):
        t1 = threading.Thread(target=self.area)
        t2 = threading.Thread(target=self.perimeter)
        t3 = threading.Thread(target=self.display_info_color)
        # run t1 and t2
        t1.start()
        t3.start()
        t2.start()
        # wait t1 and t2 before print finish task
        t1.join()
        t3.join()
        print("finished tasks")


# make object circle
circle = Circle(radius=5, color="pink")

print("~~~~~~~~~~~~display_info_shape~~~~~~~~~~~~")
# call method must make object first
circle.display_info_shape()
print("~~~~~~~~~~~~display_info_color~~~~~~~~~~~~")
Circle.display_info_color()
circle.display_info_color()
print(f"before: {circle.radius}")
print(f"set radius to 10")
circle.radius = 10
print(f"after: {circle.radius}")
circle.display_info_shape()

print("~~~~~~~~~~~~is_valid_color~~~~~~~~~~~~")
# call directly without make object
Shape.is_valid_color(circle.color)
circle.is_valid_color(circle.color)
print("~~~~~~~~~~~~thread~~~~~~~~~~~~")
# will print
circle.run()
```
